```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
```

```python
nltk.data.load('tokenizers/punkt/english.pickle')
```

```python
nltk.download()
```

```python
text = open('Resume.txt').read()
text
```

```python
def preprocessing(text):
    words = word_tokenize(text)
    tokens = [w for w in words if w.lower() not in string.punctuation]
    stopw = stopwords.words('english')
    tokens = [token for token in tokens if token not in stopw]
    # remove words less than three letters
    tokens = [word for word in tokens if len(word)>=3]
    # lemmatize
    lemma = WordNetLemmatizer()
    tokens = [lemma.lemmatize(word) for word in tokens]
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text 
```

```python
text = preprocessing(text)
text
```

```python
wordcloud = WordCloud().generate(text)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wordcloud.recolor(random_state=2017))
plt.title('Most Frequent Words')
plt.axis("off")
plt.show()
```

```python
from PIL import Image
import numpy as np
```

```python
im = Image.open("professional_me.jpg")
im.show()
```

```python
_mask = np.array(im)
```

```python
wc = WordCloud(background_color="pink", max_words=500, mask=_mask,colormap='Dark2')
# generate word cloud
wc.generate(text)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.imshow(_mask, cmap=plt.cm.Dark2, interpolation='bilinear',alpha=0.5)
plt.axis("off")
plt.figtext(.5,.9,'MY RESUME AS A WORD CLOUD', fontsize=100, ha='center')
plt.show()
```
